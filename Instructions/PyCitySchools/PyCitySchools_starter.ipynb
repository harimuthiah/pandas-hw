```python
# Dependencies and Setup
import pandas as pd
import functools 
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
tot_school = school_data["School ID"].count()
tot_student = student_data["Student ID"].count()
tot_budget = school_data["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()
tot_passing_math = student_data.loc[student_data["math_score"] >= 70]["Student ID"].count()
avg_passing_math = (tot_passing_math/tot_student)
percent_passing_math = avg_passing_math*100
tot_passing_reading = student_data.loc[student_data["reading_score"] >= 70]["Student ID"].count()
avg_passing_reading = (tot_passing_reading/tot_student)
percent_passing_reading = avg_passing_reading*100 
overall_passing_rate = ((avg_passing_math + avg_passing_reading) / 2)*100

District_Summary_df = pd.DataFrame({
    "Total Schools": [tot_school],
    "Total Students": [tot_student],
    "Total Budget": [tot_budget],
    "Average Maths Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [percent_passing_math], 
    "% Passing Reading": [percent_passing_reading],  
    "% Overall Passing Rate": [overall_passing_rate]
})
District_Summary_df["Total Students"] = District_Summary_df["Total Students"].apply('{:,}'.format)
District_Summary_df["Total Budget"] = District_Summary_df["Total Budget"].apply('${:,.2f}'.format)
District_Summary_df
```

```python
aggregations = {
    'reading_score': {
        'Average Reading Score': 'mean',
        'Total Passing Reading': lambda x: ((x >= 70).sum())
    }, 
    'math_score': {
        'Average Math Score': 'mean',
        'Total Passing Math': lambda x: ((x >= 70).sum())
    }
}
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
grades = student_data.grade.unique()
def my_merge(a,b):
    merge_df = pd.merge(a, b, how="outer", on="school_name")
    return merge_df
```

```python
dataframes = []
for i in grades:
    grd = i
    temp = student_data.loc[student_data['grade'] == i]
    temp = temp[['school_name','math_score']]
    temp.rename(columns={'math_score':grd}, inplace=True)
    temp = temp.groupby(['school_name']).mean().astype(float)
    temp.reset_index()
    #dataframes["dataframe_"+str(i)] = temp
    dataframes.append(temp)
    
merge_df = functools.reduce(my_merge, dataframes)
```

```python
merge_df.head()
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
dataframes = []
for i in grades:
    grd = i
    temp = student_data.loc[student_data['grade'] == i]
    temp = temp[['school_name','reading_score']]
    temp.rename(columns={'reading_score':grd}, inplace=True)
    temp = temp.groupby(['school_name']).mean().astype(float)
    temp.reset_index()
    #dataframes["dataframe_"+str(i)] = temp
    dataframes.append(temp)
    
merge_df = functools.reduce(my_merge, dataframes)
merge_df.head()
```
